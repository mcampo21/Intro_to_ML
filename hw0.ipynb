```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D3.csv')
df.head() #Get first rows from the dataset
```

```python
# Create Variable arrays of [1 x m]

x1 = df.values[:,0]
x2 = df.values[:,1]
x3 = df.values[:,2]
Y = df.values[:,3]
m = len(Y)
print('x1 = ', x1[: 5])
print('x2 = ', x2[: 5])
print('x3 = ', x3[: 5])
print('Y = ', Y[: 5])
print('m = ', m)
```

```python
#Single column of ones
X_0 = np.ones((m,1))
```

```python
#Reshape to convert array into 2D array of dimensions [m x 1]
X_1 = x1.reshape(m,1)
X_2 = x2.reshape(m,1)
X_3 = x3.reshape(m,1)
```

```python
# stack X_0 and X_n horizontally
X1 = np.hstack((X_0, X_1))
X2 = np.hstack((X_0, X_2))
X3 = np.hstack((X_0, X_3))
X_total = np.hstack((X_0, X_1, X_2, X_3))
```

```python
def cost_function(x, y, theta):
    
    prediction = x.dot(theta)
    error = np.subtract(prediction, y)
    sqrError = np.square(error)
    J = 1/(2*m) * np.sum(sqrError)
    
    return J
```

```python
def gradient_descent(x, y, theta, alpha, iterations):
    
    loss_history = np.zeros(iterations)
    
    for i in range(iterations):
        prediction = x.dot(theta)
        error = np.subtract(prediction, y)
        G = (1/m) * x.transpose().dot(error);
        theta = theta - (G * alpha);
        loss_history[i] = cost_function(x, y, theta)
        
    return theta, loss_history
```

```python
# Gradient Descent for X1
theta1 = np.zeros(2) 
iterations1 = 1000;
alpha = 0.01;
theta1, loss_history1 = gradient_descent(X1, Y, theta, alpha, iterations1)
print('Final value of theta =', theta1)
print('history = ', loss_history1[iterations1 -1])
```

```python
# Gradient Descent for X2
theta2 = np.zeros(2) 
iterations2 = 150;
alpha = 0.02;
theta2, loss_history2 = gradient_descent(X2, Y, theta, alpha, iterations2)
print('Final value of theta =', theta2)
print('history = ', loss_history2[iterations2 -1])
```

```python
# Gradient Descent for X3
theta3 = np.zeros(2) 
iterations3 = 1200;
alpha = 0.01;
theta3, loss_history3 = gradient_descent(X3, Y, theta, alpha, iterations3)
print('Final value of theta =', theta3)
print('history = ', loss_history3[iterations3 -1])
```

```python
plt.scatter(X1[:,1], Y, color='red', marker= '+', label= 'Training Data')
plt.plot(X1[:,1], X1.dot(theta1), color='green', label='Linear Regression')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Linear Regression of X1')
plt.legend()
```

```python
plt.plot(range(1,iterations1+1), loss_history1, color='red')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for X1')
```
