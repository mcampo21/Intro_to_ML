## Linear Regression

```python
from sklearn import metrics

def print_evaluate(true,predicted):
    mae = metrics.mean_absolute_error(true,predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true,predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('\nMAE:', mae)
    print('\nMSE:', mse)
    print('\nRMSE:', rmse)
    print('\nR2 Square', r2_square)
```

```python
# Create the linear model
# Define the Model
lr = LinearRegression()
# Fit the Model
lr.fit(train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
#Predict the model
y_pred = lr.predict(test)
```

```python
print("Error/Accuracy Analysis :")
print_evaluate(y_test, y_pred)
```

```python
def cost_function(x, y, theta):
    
    prediction = x.dot(theta)
    error = np.subtract(prediction, y)
    sqrError = np.square(error)
    J = 1/(2*m) * np.sum(sqrError)
    
    return J
```

```python
def gradient_descent(x, y, theta, alpha, iterations):
    
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        prediction = x.dot(theta)
        error = np.subtract(prediction, y)
        G = (1/m) * x.transpose().dot(error);
        theta = theta - (G * alpha);
        
        cost_history[i] = cost_function(x, y, theta)

    return theta, cost_history
```

```python
# Reshape the array
tr = len(train)
ts = len(test)
tr_one = np.ones((tr,1))
ts_one = np.ones((ts,1))
train_p1a = np.hstack((tr_one, train))
test_p1a = np.hstack((ts_one, test))
train_p1a.shape
```

```python
# Gradient Descent for p1a training data
theta = np.zeros(18)
iterations1a = 2000;
alpha = 0.002;
m = len(train_p1a)
theta_p1a_train, cost_history1a_train = gradient_descent(train_p1a, y_train, theta, alpha, iterations1a)
m = len(test_p1a)
theta_p1a_test, cost_history1a_test = gradient_descent(test_p1a, y_test, theta, alpha, iterations1a)

print('Final value of train theta =', theta_p1a_train)
print('Train cost history = ', cost_history1a_train[iterations1a -1])
print('Final value of test theta =', theta_p1a_test)
print('Test cost history = ', cost_history1a_test[iterations1a -1])
```

```python
# Plot gradient descent of p1a
plt.plot(range(1,iterations1a+1), cost_history1a_train, color='red');
plt.plot(range(1,iterations1a+1), cost_history1a_test, color='blue');
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('Number of iterations');
plt.ylabel('Cost (J)');
plt.title('Convergence of gradient descent for Training data');
```

# PCA Analysis with Linear Regression

```python
# Implement PCA
from sklearn.decomposition import PCA
pca_train = PCA(n_components=2)
pca_test = PCA(n_components=2)
train_pca = pca_train.fit_transform(train)
test_pca = pca_test.fit_transform(test)
```

```python
# Create PCA Linear Regression Model
lr.fit(train_pca, y_train)
```

```python
# Find the N value that gives the lowest RMSE
#%%capture
pca_rmse = np.zeros(17)
pca_r2 = np.zeros(17)
n = 0
best = 1

for i in range(1,18):
    pca_train = PCA(n_components=i)
    pca_test = PCA(n_components=i)
    trainData_pca = pca_train.fit_transform(train)
    testData_pca = pca_test.fit_transform(test)
    lr.fit(trainData_pca, y_train);
    Y_pred_pca = lr.predict(testData_pca)
    rmse = np.sqrt(metrics.mean_squared_error(y_test, Y_pred_pca))
    r2_square = metrics.r2_score(y_test, Y_pred_pca)
    pca_rmse[i-1] = rmse
    pca_r2[i-1] = r2_square
    
    if (rmse < best):
        best = rmse
        n = i
```

```python
print('The best N-component for PCA is ', n)

# plot rmse
plt.plot(range(1,18), pca_rmse, color='red', label="RMSE");
#plt.plot(range(1,18), pca_r2, color='blue', label="R2_Squared");
#plt.plot(range(1,31), pca_recall, color='green', label="recall");
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('n-Components');
plt.ylabel('Percentage');
plt.title('Values per Princple Components');
plt.legend(loc= "upper right");
```
